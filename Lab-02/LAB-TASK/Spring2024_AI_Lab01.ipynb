### Properties of Array
* __Size__ Returns the size (number of elements) of an array
* __ndim__ Returns the dimensions (1D, 2D etc) of an array
* __shape__ Returns the shape (rows, cols) of an array

```python
print(a)
print(a.size)
print(a.ndim)
print(a.shape)
```

### Two Dim Arrays and it properties

```python
b = np.array([ [1,2],
               [3,4],
               [5,6] ])

print(b)
print(b.size)
print(b.ndim)
print(b.shape)
```

### Reshaping the array
* __reshape()__ function can be used to reshape/redfine the number of rows and cols

```python
c = np.array([ 1,12,3,44,5,6,67,8])

d = c.reshape(4,2)

print("Array c : ", c)
print("Array d : ", d)
print("c.shape : ", c.shape)
print("d.shape : ", d.shape)
```

### Array elements can be modified

```python
c = np.array([ 1,2,3,4,5,6,7,8])

print("Before change in Array c : ", c)
c[0] = 100
print("After change in Array c : ", c)
```

### Slicing the array

```python
c = np.array([1,2,3,4,5,6,7,8])

d =  c[1:4].copy()

print(d)
```

```python
c[3:5] = 40,50

print(c)

print(d)
```

```python

# Generate data for the first line
x1 = np.linspace(0, 10, 20)
y1 = np.sin(x1)

# Generate data for the second line
x2 = np.linspace(0, 10, 20)
y2 = np.cos(x2)

# Create a figure and axes
fig, ax = plt.subplots()

# Plot the first line with a circle marker
ax.plot(x1, y1, '-o', label='sin')

# Plot the second line with a diamond marker
ax.plot(x2, y2, '-D', label='cos')


# Add a legend to the plot
ax.legend()

# Set the title and labels
ax.set_title("Sine and Cosine")
ax.set_xlabel("x")
ax.set_ylabel("y")

# Show the plot
plt.show()
```

## NetworkX

```python
import networkx as nx
```

```python
# Create an empty graph
G = nx.Graph()
```

```python
# Adding nodes manually
G.add_node(1)
G.add_node(2)
G.add_node(3)
G.add_node(4)
```

```python
G.add_nodes_from([5,6])
```

```python
# Adding an edge
G.add_edge(1,2)
G.add_edge(1,3)
G.add_edge(3,1)
G.add_edge(2,4)
G.add_edge(4,1)
G.add_edge(4,5)
G.add_edge(5,6)
```

```python
# Drawing graph
nx.draw(G)
```

```python
# Adding edge between non-existent nodes
G.add_edge(1,9)
```

```python
# Adding edges from a list
G.add_edges_from ([(1,4), (2,5)])
```

```python
# Drawing graph
nx.draw(G)
```

```python
# Drawing options
options = {
    "font_size": 10,
    "node_size": 500,
    "node_color": "white",
    "edgecolors": "blue",
    "linewidths":4,
    "width": 4,
}
```

```python
# Drawing with options
nx.draw_networkx(G, **options)
```

```python
# Writing graph to a text file
nx.write_edgelist(G,"graph.txt")
```
